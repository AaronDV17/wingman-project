## Data Merging

```python
def merge_dataframes(df_ref, other_dfs, index_column):
    for df in other_dfs:
        df_ref = df_ref.merge(df, how='inner', on=index_column)
        
    return df_ref
```

- 'phase_no' & 'eventsoe_no' == y targets
- Therefore other tables (exc. Findings_df) are inner joined on 'id' column
