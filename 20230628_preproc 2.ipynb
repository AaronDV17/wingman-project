# Baseline Preprocessing

## Imports and Data

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import LabelEncoder
```

```python
wingman_data = pd.read_csv('raw_data/wingman_data.csv')
wingman_data.set_index('id', inplace=True)
```

```python
wingman_data.dtypes
```

```python
wingman_data
```

## Encoding Categorical Features

```python
wingman_cat = wingman_data.select_dtypes(include=['object'])
wingman_cat.head()
```

```python
# One Hot Encoding wingman_cat

ohe = OneHotEncoder(sparse_output=False).fit(wingman_cat)
wingman_cat_trans = pd.DataFrame(ohe.transform(wingman_cat))
wingman_cat_trans
```

```python
wingman_cat_trans = pd.DataFrame(wingman_cat_trans)
wingman_cat_trans
```

## Scaling Numerical Features

```python
wingman_num = wingman_data.select_dtypes(include=['int64', 'float64'])
wingman_num.head()
wingman_num = wingman_num.drop(columns='eventsoe_no')
wingman_num
```

```python
# Scaling wingman_num

scaler = RobustScaler().fit(wingman_num)
wingman_num_trans = pd.DataFrame(scaler.transform(wingman_num))


```

## Preprocessed Data

```python
## Concatenating categorical and numerical data

X_preproc = pd.concat([wingman_cat_trans, wingman_num_trans], axis=1)
```

```python
## Encoding the target

y = wingman_data['eventsoe_no']
y = y.astype("string")

label_encoder = LabelEncoder()

label_encoder.fit(y)

y_enc = label_encoder.transform(y)

y_enc = pd.DataFrame(y_enc)
```
